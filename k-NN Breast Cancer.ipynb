```python
#load breast cancer dataset
from sklearn.datasets import load_breast_cancer
# numerical ops and handling array
import numpy as np
#manip dataframes (or data manip and analysis)
import pandas as pd
#creating static interactive / or animated visulaizations
import matplotlib.pyplot as plt #plotting

#1. DATASET OVERVIEW
data = load_breast_cancer() #loading the breast cancer dataset to data
df=pd.DataFrame(data.data,columns=data.feature_names) #putting features name and data (2D) of breast cancer to dataframe of pandas
print(df.head().to_string()) #render df.head to a consule tabular output

#2. DATA SCATTER PLOT
rand_select = np.random.choice(df.columns, size=2, replace=False) #using numpy for selecting 2 non-duplicated columns
x_column , y_column = rand_select #unpacking the two names for easier access
plt.figure(figsize=(6,3))#creating a 6x3 figure (inches)
plt.scatter(df[x_column],df[y_column],alpha=0.7,edgecolors='k') #creating scatter plot of x and y with transpareny of .7 and edge color of data point black
plt.xlabel(x_column)#labeling x axis
plt.ylabel(y_column)#labeling y axis as y
plt.show()

```
